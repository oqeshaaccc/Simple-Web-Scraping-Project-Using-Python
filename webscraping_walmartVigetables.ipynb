```python
import requests
from bs4 import BeautifulSoup
import urllib.request
from PIL import Image 


# Make the request and get the response content
url = "https://www.walmart.ca/ce/grocery/weekly-flyer-features/6000196190101?c=10019+6000194327370&icid=OGHP_TC_Flyer_WM"


# Send a GET request to the URL
response = requests.get(url)

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the product items on the page
product_items = soup.find_all("div", class_="e1m8uw9118")
#print (product_items)
# Loop through each product item and extract the image, title, and link
for product in product_items:
    # Extract the image URL
    image = product.find("img")["src"]
    # Extract the product title
    title = product.find("div", class_="e1m8uw914").text.strip()
    # Extract the product link
    link = "https://www.walmart.ca" + product.find("a")["href"]
     
    # Extract the price
    #price = abc.find("span", {"data-automation": "current-price"})
    # Extract the volume
    volume = product.find("div", class_="e1m8uw915").text.strip()

    # Print the product information
    urllib.request.urlretrieve(image ,"gfg.png")
    img = Image.open("gfg.png")
    img.show()
    print("Title: ", title)
    #print(f"Price: {price}")
    print("Volume: ", volume)
    print("Link: ", link)
    print("Image: ", image)
    print("\n")
print ("################ End ################")
```
